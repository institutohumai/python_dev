En este ejercicio tenemos un codigo con mucho para mejorar! es tu trabajo como desarrollador identificar que es lo que esta mal con este codigo.

1- Implementa type hints

2- Documentacion basica que describa que es lo que esta haciendo este codigo

3- Identifica cuales son las buenas practicas que no esta siguiendo

4- Lo mas importante, reescribe este codigo siguiendo las buenas practicas! hay mucho por hacer!

```python
def calculate_average_above_treshould(values, treshold):
    temp = 0
    k = 0
    if treshold is not None and values is not None:
        for value in values:
            if value > treshold:
                temp += 1
                k += 1
            if k == 0:
                return None
            else:
                average = temp / k
                return average
    elif values is None:
        raise ValueError("List is None")
    elif len(values) == 0:
        raise ValueError('Empty list')
    else:
        raise ValueError('Treshold is None')
    return None
```
